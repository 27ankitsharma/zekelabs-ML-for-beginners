### Variance Threashold
* Dimensionality Reduction

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
var_threash = VarianceThreshold(threshold=.5)
```

```python
import numpy as np
```

```python
var_threash.fit_transform(iris.data)
```

### Removing of Irrelevent features for classification

```python
from sklearn.feature_selection import SelectKBest, chi2, f_classif, f_regression
```

```python
features = iris.data
target = iris.target
```

```python
chi2_selector = SelectKBest(chi2, k =2)
```

```python
chi2_selector.fit_transform(features,target)
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
from sklearn.svm import LinearSVC
```

```python
svc = LinearSVC().fit(iris.data, iris.target)
```

```python
model = SelectFromModel(svc)
```

```python
model.fit_transform(iris.data, iris.target)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
X = pca.fit_transform(iris.data)
```

```python
svc.fit(X,iris.target)
```

```python
svc.predict(X)
```

```python
pca = PCA(n_components=.9, whiten=True)
```

```python
np.var(iris.data)
```

```python
d = pca.fit_transform(iris.data)
```
