```python
import pandas as pd
```

```python
ratings = pd.read_csv('Data/ml-100k/u.data',sep='\t',names=['user_id','movie_id','rating'], usecols=range(3))
```

```python
movies = pd.read_csv(r'Data/ml-100k/u.item', sep='|',encoding='latin-1', names=['movie_id','title'], usecols=range(2))
```

```python
ratings = pd.merge(movies,ratings)
```

```python
ratings.head()
```

```python
movieRatings = ratings.pivot_table(index=['user_id'], columns=['title'], values='rating')
```

```python
movieRatings.head()
```

```python
starWarsRating = movieRatings['Star Wars (1977)']
```

```python
starWarsRating.head()
```

### Corrwith

```python
similarMovies = movieRatings.corrwith(starWarsRating)
```

```python
similarMovies = similarMovies.dropna()
```

```python
df = pd.DataFrame(similarMovies)
```

```python
df.head(10)
```

```python
movieRatings[['Star Wars (1977)',"'Til There Was You (1997)"]].head()
```

```python
import numpy as np
```

```python
movieStats = ratings.groupby('title').agg({'rating':[np.size, np.mean]})
```

```python
popularMovies = movieStats['rating']['size'] >= 100
```

```python
movieStats[popularMovies].sort_values([('rating','size')], ascending=False)
```
