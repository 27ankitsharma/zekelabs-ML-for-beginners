Group observations using hierchy of clusters

```python
from sklearn.cluster import AgglomerativeClustering, KMeans
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(iris.data[:,:2])
```

```python
hcluster = AgglomerativeClustering(n_clusters=3)
```

```python
hcluster.fit(X)
```

```python
hcluster.labels_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(iris.data[:,0], iris.data[:,1], c=hcluster.labels_)
```

```python
km = KMeans(n_clusters=3)
```

```python
km.fit(X)
```

```python
km.labels_
```

```python
plt.scatter(iris.data[:,0], iris.data[:,1], c=km.labels_)
```
