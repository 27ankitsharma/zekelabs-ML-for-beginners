```python
from sklearn.datasets import make_circles
```

```python
X,Y = make_circles(n_samples=200, noise=.05)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0],X[:,1],c = Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
models = []
for degree in range(1,6):
    pol = PolynomialFeatures(degree)
    X_tf = pol.fit_transform(X)
    trainX,testX,trainY,testY = train_test_split(X_tf,Y)
    lr = LogisticRegression(class_weight="")
    lr.fit(trainX,trainY)
    models.append(lr)
    print('Degree', degree,  lr.score(testX,testY))
```

```python
pol = PolynomialFeatures(2)
X_tf = pol.fit_transform(X)
#trainX,testX,trainY,testY = train_test_split(X_tf,Y)
o = models[1].predict(X_tf)
```
