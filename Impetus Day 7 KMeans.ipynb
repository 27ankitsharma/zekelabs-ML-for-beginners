```python
import pandas as pd
```

```python
rental_data = pd.read_csv('data/house_rental_data.csv', index_col='Unnamed: 0')
```

```python
rental_data.info()
```

```python
from sklearn.cluster import KMeans
```

```python
scores = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(rental_data)
    scores.append(-1 * kmeans.score(rental_data))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(scores)
```

```python
kmeans = KMeans(n_clusters=3)
```
