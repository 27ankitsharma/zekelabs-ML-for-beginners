```python
df = pd.DataFrame({'c1':[11,12,3], 'c2':[4,5,6]})
```

```python
def f(r):
    if r > 10:
        return 'Big'
    else:
        return 'Small'
df.c1.apply(f)
```

```python
def f(r):
    if r['c1'] > r['c2']:
        return r['c1']
    else:
        return r['c2']
    
df.apply(f, axis=1)
```

```python
df.apply(lambda r: r['c1'] if r['c1'] > r['c2'] else r['c2'], axis=1)
```

```python
df.apply(lambda x:sum(x), axis=0)
```

### Add gender before name

```python
def gadd(r):
    return r['Sex'] + ': ' + r['Name']
    

titanic_data['gender-name'] = titanic_data.apply(gadd, axis=1)
```

```python
titanic_data.head()
```

```python
import numpy as np

def gadd(r):
    print (r)
    print ('*')
    return (np.sum(r == 1), np.sum(r==2), np.sum(r == 3))

titanic_data.Pclass.apply(gadd)
```

#### groupBy

```python
gb = titanic_data.groupby('Pclass')
```

```python
gb.Age.max()
```

```python
gb.Age.median()
```

```python
gb.Age.agg(['sum','max'])
```

```python

gb.agg({'Age':['sum','max','min'], 'Fare':['sum','min','max']})
```

```python
def g(d):
    return (np.mean(d.Fare))

gb.apply(g)
```
