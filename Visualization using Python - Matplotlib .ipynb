#### Basic Plotting : plot
* Series & DataFrame functions have plot function in it

```python
import pandas as pd
import numpy as np
```

```python
ts = pd.Series(np.arange(1000), index= pd.date_range('1/1/2000',periods=1000))
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
```

```python
df = df.cumsum()
df.plot()
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
```

```python
df3.head()
```

```python
df3['A'] = pd.Series(list(range(len(df))))
```

```python
df3.head()
```

```python
df3.plot(x='A', y='B')
```

#### Histograms

```python
df4 = pd.DataFrame({'a': np.random.randn(1000) + 1, 
                    'b': np.random.randn(1000), 
                    'c': np.random.randn(1000) - 1}, 
                   columns=['a', 'b', 'c'])
```

```python
df4.plot.hist(alpha=0.5)
```

```python
df4.plot.hist(stacked=True, bins=20)
```

```python
df4.plot.hist(stacked=True, bins=20, orientation='horizontal')
```

```python
df.diff().hist(color='r', alpha=1, bins=50)
```

#### Scatter Plot

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
```

```python
df.plot.scatter(x='a',y='b')
```

```python
ax = df.plot.scatter(x='a', y='b', color='DarkBlue', label='Group 1')
df.plot.scatter(x='c', y='d', color='DarkGreen', label='Group 2', ax=ax)
```
